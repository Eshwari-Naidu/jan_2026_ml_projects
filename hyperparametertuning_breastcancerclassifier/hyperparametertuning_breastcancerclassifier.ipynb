```python

# =============================
# Breast Cancer Classification using K-Nearest Neighbors (KNN)
# =============================

# 1. Import libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
# -----------------------------
# 2. Load the dataset
# -----------------------------
breast_cancer_data = load_breast_cancer()

# Explore dataset structure
print("First data point:\n", breast_cancer_data.data[0])
print("\nFeature names:\n", breast_cancer_data.feature_names)
print("\nTarget values:\n", breast_cancer_data.target)
print("\nTarget names:\n", breast_cancer_data.target_names)

# -----------------------------
# 3. Train-validation split
# -----------------------------
training_data, validation_data, training_labels, validation_labels = train_test_split(
    breast_cancer_data.data,
    breast_cancer_data.target,
    test_size=0.2,
    random_state=100
)

print("\nTraining samples:", len(training_data))
print("Training labels:", len(training_labels))

# -----------------------------
# 4. Model training for different k values
# -----------------------------
k_list = []
accuracies = []

for k in range(1, 101):
    classifier = KNeighborsClassifier(n_neighbors=k)
    classifier.fit(training_data, training_labels)
    k_list.append(k)
    accuracies.append(classifier.score(training_data, training_labels))

# -----------------------------
# 5. Plot k vs accuracy
# -----------------------------
plt.figure(figsize=(10, 6))
plt.plot(k_list, accuracies)
plt.xlabel('Value of k')
plt.ylabel('Training Accuracy')
plt.title('K Value vs Training Accuracy')
plt.grid()
plt.show()
```
