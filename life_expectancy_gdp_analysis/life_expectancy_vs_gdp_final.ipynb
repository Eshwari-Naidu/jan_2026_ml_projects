# Life Expectancy vs GDP Analysis
This project explores the relationship between GDP and life expectancy across six countries using data from the World Health Organization and the World Bank.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df= pd.read_csv(r"all_data.csv")
print(df.head())
df.describe()
#print(df.dtypes)
```

```python
print(df.isnull().sum())
df=df.rename(columns={"Life expectancy at birth (years)" : "Life_expectancy"})
print(df.Country.unique())
#print(df.Year.unique())
```

## Research Question
Is there a relationship between a country's GDP and its life expectancy?


```python
avg_life= (df.groupby('Country')["Life_expectancy"].mean())
avg_life= avg_life.sort_values(ascending=False)
colors=["blue", "red","purple","green","darkblue", "maroon"]
print(avg_life)
avg_life.plot(kind="bar", color=colors)
plt.xlabel("Countries")
plt.ylabel("Average Life Expectancy")
plt.tight_layout()
plt.show()
plt.clf()
sns.set_style("whitegrid")
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x='Year', y='Life_expectancy', hue='Country')
plt.title("Life Expectancy Over the Years")
plt.xlabel("Years")
plt.ylabel("Life Expectancy (In Years) ")
plt.tight_layout()
plt.show()
```

### Observation

Life expectancy has generally increased over time across all six countries, though the rate of improvement differs between countries

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(
    data=df,
    x='GDP',
    y='Life_expectancy',
    hue='Country'
)

plt.xscale('log')
plt.title('GDP vs Life Expectancy (Log Scale)')
plt.xlabel('GDP (log scale)')
plt.ylabel('Life Expectancy (years)')
plt.legend(title='Country')
plt.show()
```

```python
correlation = df['GDP'].corr(df['Life_expectancy'])
correlation
```

### Correlation Analysis

The correlation coefficient between GDP and life expectancy is positive, indicating that countries with higher GDP tend to have higher life expectancy. However, correlation alone does not imply causation.


## Conclusion

This analysis shows that while GDP and life expectancy are positively related, increases in GDP do not always result in proportional gains in life expectancy. This suggests that factors beyond economic growth, such as healthcare access and social policies, play a significant role in determining population health.
